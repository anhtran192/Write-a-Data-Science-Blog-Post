```python
# Count the occurrences of each property type and sort them in descending order
property_type_counts = listings_df['property_type'].value_counts().reset_index()
property_type_counts.columns = ['Property Type', 'Count']
property_type_counts = property_type_counts.sort_values(by='Count', ascending=False)

# Plot the most common property types
plt.figure(figsize=(12, 6))
sns.barplot(x='Property Type', y='Count', data=property_type_counts.head(10))
plt.title('Top 10 Most Common Property Types in Seattle')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
